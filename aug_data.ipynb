```python
# interpreter : conda activate scipy
import numpy as np
import imageio, elasticdeform
import matplotlib.pyplot as plt
```

```python
X = numpy.zeros((270, 270, 1))
X[::10, ::10] = 1
X.shape
```

```python
plt.imshow(X)
plt.show()
```

```python
# apply deformation with a random 3 x 3 grid
X_deformed = elasticdeform.deform_random_grid(X.reshape(270,270), sigma=1, points=3, order=3, mode='constant', cval=0.0,
                       crop=None, prefilter=False, axis=None,
                       affine=None, rotate=None, zoom=1.01)
print(X_deformed.shape)
plt.imshow(X_deformed)
plt.show()
```

```python
X_deformed.reshape(270, 270, 1)[7:-7, 7:-7].shape
```

```python
imageio.imsave('test_X.png', X)
imageio.imsave('test_X_deformed.png', X_deformed)
```
